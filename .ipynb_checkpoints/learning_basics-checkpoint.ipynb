```python
# A list representing the sizes of our database tables in Gigabytes
table_sizes = [12.5, 45.1, 8.2, 102.4, 3.0]

# 1. Count how many tables we have
total_tables = len(table_sizes)

# 2. Calculate the total storage space they take up
total_storage = sum(table_sizes)

# 3. Print the results out beautifully
print("Total number of tables:", total_tables)
print("Total storage used (GB):", total_storage) 
```
